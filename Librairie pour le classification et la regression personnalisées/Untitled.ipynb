```python
from ML import config, experimentation
import pandas as pd
import numpy as np
from sklearn import datasets
```

# Pour la classification

```python
iris = datasets.load_iris()
columns = iris["feature_names"]
columns.append("iris")
data = pd.DataFrame(data = np.concatenate((iris["data"], np.reshape(iris["target"], (-1, 1))),axis=1 ) , columns = columns)
data.to_excel("iris_dataset.xlsx")
```

```python
def load_dataset(filepath: str, **kwargs):
    
    df = pd.read_excel(io = filepath, sheet_name = 'Sheet1', keep_default_na=False, index_col = 0)    
    
    return df 
```

```python
target_col = 'iris'
filepath = './iris_dataset.xlsx'
task = 'classif'
config_dir  = './experiment'
metrics = ['Accuracy', 'balancedAcc','F1']
cross_validation = ['stratified_kfold']

models = ["random_forest"]
experiment = '%s_model_selection'%target_col
conf = config.Config(target_col=target_col, filepath=filepath, task=task,experiment=experiment, config_dir=config_dir, cross_validation=cross_validation, metrics = metrics, models = models)
result_filepath =  experimentation.run_experiment(conf, load_dataset)
```

```python
resultat = experimentation.display_result(result_filepath)
drop = [col for col in list(resultat) if 'rank' in col or 'fit' in col or 'time' in col or 'split' in col]
resultat = resultat.drop(drop, axis = 1)
top = resultat.sort_values( 'mean_test_balancedAcc',ascending = False)[:10]
top
```
